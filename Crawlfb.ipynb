```python
from facebook_scraper import get_posts
import pandas as pd
```

```python
post_list = []
for post in get_posts(page,
                    options={"comments": True, "reactions": False, "allow_extra_requests": True},
                    extra_info=True, pages=10, cookies=cookie_path):
    post_list.append(post)
```

```python
post_df = pd.DataFrame(columns = [])
for post in post_list:
    post_entry = post
    fb_post = pd.DataFrame.from_dict(post_entry, orient = 'index')
    fb_post = fb_post.transpose()
    post_df = post_df.append(fb_post)
    print(post['post_id'] + 'get')
```

```python
keys_drop = ['comment_url', 'commenter_id', 'commenter_url']

for i in range(len(post_list)):
    for comment in post_list[i]['comments_full']:
        for key in keys_drop:
            del comment[key]
```

```python
key = 'comment_text'
comments_only = []
for i in range(len(post_list)):
    for comment in post_list[i]['comments_full']:
        comments_only.append(comment['comment_text'])
len(comments_only)
```

```python
key = 'replies'
keys = []
for i in range(len(post_list)):
    for comment in post_list[i]['comments_full']:
        keys.append(comment['replies'])
len(keys)
```

```python
keys_drop = ['comment_url', 'commenter_id', 'commenter_url', 'comment_id','commenter_name', 'commenter_meta', 'comment_time', 'comment_image', 'comment_reactors', 'comment_reactions', 'comment_reaction_count']
for i in range(len(keys)):
    for com in keys[i]:

      for key in keys_drop:
          del com[key]
```

```python
replies_only = []
for i in range(len(keys)):
    for comment in keys[i]:

        replies_only.append(comment['comment_text'])
len(replies_only)
```

```python

not_drop = ['text']
df_fb = post_df[not_drop]
df_fb.info()
```

```python
com = pd.DataFrame(comments_only)
type(com)
```

```python
rep = pd.DataFrame(replies_only)
type(rep)
```

```python
new = pd.concat([com,rep],ignore_index = True, axis = 1)
new.head()
```

```python
new.columns = ["Comments", "Replies"]
new.head()
```

```python
# Saving to csv
path = content_path + ".csv"
df_fb.to_csv(content_path  + ".csv", index = False)
print(path)
```
